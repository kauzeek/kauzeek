```python
team_runs = data.groupby('team')['runs'].sum().sort_values(ascending=False)
team_runs
```

```python
import pandas as pd
fig, ax = plt.subplots()
plt.bar(team_runs.index, team_runs.values)
plt.xlabel('Teams')
plt.ylabel('Total Runs Scored')
plt.title('Total Runs Scored by Each Team')
plt.show()
```

```python
data["bat_or_bowl"]
```

```python
batting_count = data[data['bat_or_bowl'] == "bat"].shape[0]
bowling_count = data[data['bat_or_bowl'] == "bowl"].shape[0]
```

```python
labels = ['Batting', 'Bowling']

sizes = [batting_count, bowling_count]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Distribution of Batting and Bowling Performances')
plt.show()
```

```python
batting_data = data[data['bat_or_bowl'] == 'bat']
top_scorers = batting_data.sort_values(by='runs', ascending=False).head(10)
print("Top 10 Batsmen by Runs Scored:")
print(top_scorers[['player', 'runs']].to_string(index=False))
```

```python
players = top_scorers['player'].to_numpy()
runs = top_scorers['runs'].to_numpy()


palette = sns.color_palette("Greens")

plt.bar(players, runs, color=palette)
plt.xlabel('Players')
plt.ylabel('Runs Scored')

plt.xticks(rotation=45)

plt.show()
```

```python
bowling_data = data[data['bat_or_bowl'] == 'bowl']
top_wicket_takers = bowling_data.sort_values(by='wkts', ascending=False).head(10)
print("Top 10 Bowlers by Wickets Taken:")
print(top_wicket_takers[['player', 'wkts']].to_string(index=False))
```

```python
players = top_wicket_takers['player'].to_numpy()
wickets = top_wicket_takers['wkts'].to_numpy()

palette = sns.color_palette("Purples")

plt.bar(players, wickets, color=palette)

plt.title('Top Wicket Takers')
plt.xlabel('Bowler')
plt.ylabel('Wickets Taken')
plt.xticks(rotation=45)


plt.show()
```

```python
team_data = bowling_data.groupby('team')['wkts'].mean()
print("\nAverage Wickets Taken per Innings (by Team):")
print(team_data.to_string())
```

```python
import matplotlib.pyplot as plt

palette = sns.color_palette("Oranges")

top_wicket_takers.plot.bar(x='team', y='wkts',color=palette)

plt.title('Top Wicket Takers')
plt.xlabel('Bowler')
plt.ylabel('Wickets Taken')
plt.xticks(rotation=45)
plt.show()
```

```python
custom_colors = sns.color_palette("tab10", 10)
plt.figure(figsize=(12, 20))
sns.barplot(x='runs', y='team', hue='opposition', data=data,dodge=False,palette=custom_colors)
plt.xlabel('Runs')
plt.ylabel('Ground')
plt.title('Team with Highest Runs on Each Ground')
plt.legend(title='Opposition', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```
